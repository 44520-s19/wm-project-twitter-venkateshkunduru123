### Retriving information from Twitter and store in pickle file:

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; In these, I collect the data from twitter and store that twitter data in the pickle files and Retrieve the data from that pickle file.

```python
import tweepy
import pickle
from twitter_keys import consumer_key, consumer_secret, access_token, access_secret

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth)
# Retriving information from account 1 and storing in pickle
account1_tweets = api.user_timeline('EmmaWatson')
with open('emma.pkl', 'wb') as f:
    pickle.dump(account1_tweets, f)

# Retriving information from account 2 and storing in pickle
account2_tweets = api.user_timeline('SunnyLeone')
with open('sunny.pkl', 'wb') as f:
    pickle.dump(account2_tweets, f)
    
# Retriving information from account 3 and storing in pickle
account3_tweets = api.user_timeline('urstrulyMahesh')
with open('mahesh.pkl', 'wb') as f:
    pickle.dump(account3_tweets, f)
```

# Lexical Diversity
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; the ratio of different unique stems to the total number of words is said to be lexical diversity

```python
def lexical_diversity(text):
    tokens = text.lower().split()
    return len(set(tokens))/len(tokens)

print(account1_tweets[0].text)
print('lexical diversity for account 1 tweet: ', lexical_diversity(account1_tweets[0].text))
print(account2_tweets[0].text)
print('lexical diversity for account 2 tweet: ', lexical_diversity(account2_tweets[0].text))
print(account3_tweets[0].text)
print('lexical diversity for account 3 tweet: ', lexical_diversity(account3_tweets[0].text))
```

# Commonly used words
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; most repeated words in a paragraph were analyzed and the results were shown in the form of histogram containing number of occurences and different words as their axes.

```python
para = "Mahesh Babu (born Mahesh Ghattamaneni on 9 August 1975) is an Indian actor, producer, media personality, and philanthropist known for his works in Telugu cinema. He owns the production house G. Mahesh Babu Entertainment Pvt. Ltd. The younger son of veteran Telugu actor Krishna, Mahesh made his cameo as a child artist in Needa (1979), at the age of four, and acted in eight other films as a child artist. He made his debut as a lead actor with Rajakumarudu (1999) and won the State Nandi Award for Best Male Debut."

from collections import Counter
words = para.lower().split()
word_count = Counter(words)
print(word_count)
```

# Device Used
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; In Twitter, the tweet is from which device(Iphone, Android and Web client)

```python
print('--------------------------------account 1-------------------------------')
account1_iphone=0
account1_android=0 
account1_web = 0
for i in account1_tweets:
    if i.source=='Twitter for iPhone':
        account1_iphone += 1
    elif i.source=='Twitter for Android':
        account1_android += 1
    elif i.source=='Twitter Web Client':
        account1_web += 1
print('Iphone:', account1_iphone, 'Android:', account1_android, 'Web Client:', account1_web)

print('--------------------------------account 2-------------------------------')
account2_iphone=0
account2_android=0
account2_web = 0
for i in account2_tweets:
    if i.source=='Twitter for iPhone':
        account2_iphone += 1
    elif i.source=='Twitter for Android':
        account2_android += 1
    elif i.source=='Twitter Web Client':
        account2_web += 1
print('Iphone:', account2_iphone, 'Android:', account2_android, 'Web Client:', account2_web)
    
print('--------------------------------account 3-------------------------------')
account3_iphone=0
account3_android=0
account3_web = 0
for i in account1_tweets:
    if i.source=='Twitter for iPhone':
        account3_iphone += 1
    elif i.source=='Twitter for Android':
        account3_android += 1
    elif i.source=='Twitter Web Client':
        account3_web += 1
print('Iphone:', account3_iphone, 'Android:', account3_android, 'Web Client:', account3_web)
```

# Visualization for Devices used
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Pictorial representation of Devices used

```python
import numpy as np
import matplotlib.pyplot as plt

N = 3
ind = np.arange(N)  # the x locations for the groups
width = 0.27       # the width of the bars

fig = plt.figure()
ax = fig.add_subplot(111)
iphone = [account1_iphone,account2_iphone,account3_iphone]
rects1 = ax.bar(ind, iphone, width, color='r')
android = [account1_android,account2_android,account3_android]
rects2 = ax.bar(ind+width, android, width, color='g')
web = [account1_web, account2_web, account3_web] 
rects3 = ax.bar(ind+width*2, web, width, color='b')

ax.set_ylabel('No of users')
ax.set_xlabel('Accounts')
ax.set_xticks(ind+width)
ax.set_xticklabels( ('Account 1', 'Account 2', 'Account 3') )
ax.legend( (rects1[0], rects2[0], rects3[0]), ('IPhone', 'Android', 'Web Client') )

def autolabel(rects):
    for rect in rects:
        h = rect.get_height()
        ax.text(rect.get_x()+rect.get_width()/2., 1.05*h, '%d'%int(h),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

plt.title('Devices used')
plt.show()
```

# Sentiment Analysis
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Sentiment Analysis is the process of ‘computationally’ determining whether a piece of writing is positive, negative or neutral. It’s also known as opinion mining, deriving the opinion or attitude of a speaker.

```python
import pickle 
from nltk_helpers import get_sentiments 
    
print(get_sentiments(account1_tweets[0].text))
print(get_sentiments(account2_tweets[0].text))
print(get_sentiments(account3_tweets[0].text))
```

# Visualization of sentiment Analysis
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 3D representation of Sentiment Analysis.

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
# build our figure
fig = plt.figure()
# build the axes
ax = fig.add_subplot(111, projection='3d')

import pickle
from nltk_helpers import split_sentiments

ac1X, ac1Y, ac1Z = split_sentiments([get_sentiments(tweet.text) for tweet in account1_tweets])
ax.scatter(ac1X, ac1Y, ac1Z, color = 'g', marker = 'd', label='Emma Watson')

ac2X, ac2Y, ac2Z = split_sentiments([get_sentiments(tweet.text) for tweet in account2_tweets])
ax.scatter(ac2X, ac2Y, ac2Z, color = 'b', marker = 'o', label='sunny Leone')

ac3X, ac3Y, ac3Z = split_sentiments([get_sentiments(tweet.text) for tweet in account3_tweets])
ax.scatter(ac3X, ac3Y, ac3Z, color = 'r', marker = 'p', label = 'Mahesh Babu')


ax.set_xlabel('Negative')
ax.set_ylabel('Neutral')
ax.set_zlabel('Positive')
ax.legend(loc='upper left', shadow = True)

plt.show()
```

# 2D representation of Sentiment Analysis
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 2D representation of Sentiment analysis(positive vs negative sentiments)

```python
plt.scatter(ac1X, ac1Y, color = 'g', marker = 'D', label = 'Emma Watson')
plt.scatter(ac2X, ac2Y, color = 'b', marker = 'o', label = 'Sunny Leone')
plt.scatter(ac3X, ac3Y, color = 'r', marker = 'P', label = 'Mahesh Babu')
plt.xlabel('Negative')
plt.ylabel('Positive')
plt.title('Positive vs Negative Sentiments')
plt.legend(loc = 'upper right', shadow = True)
plt.show()
```
